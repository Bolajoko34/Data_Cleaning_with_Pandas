## Cleaning Different Sheets from the Same File in Python

```python
# import pandas
import pandas as pd
```

### Load dataset

```python
financial_inclusion = pd.read_excel(
    r"c:\Users\TOSHIBA\Desktop\INTERNSHIP\Financial Inclusion Dataset.xlsx",
    sheet_name=["Money Supply", "Mobile Money","Interest Rate"],
    engine="openpyxl"
)
```

```python
money_supply = financial_inclusion["Money Supply"]
mobile_money = financial_inclusion["Mobile Money"]
interest_rate = financial_inclusion["Interest Rate"]
```

### Preview

```python
money_supply.head()
```

### intia Inspection

```python
# check structure
money_supply.info()

# check summary stats
money_supply.describe()

# check missing values
money_supply.isnull().sum()
```

### Clean column names

```python
money_supply.rename(columns ={
    "Variables":"Variable",
    "Jan":"January",
    "Feb":"Febuary",
    "Mar":"March",
    "Apr":"April",
    "Jun":"June",
    "Jul":"July",
    "Aug":"August",
    "Sep":"September",
    "Oct":"October",
    "Nov":"November",
    "Dec":"December"}, inplace =True)
```

```python
money_supply 
```

### Remove (Millions of Ghana cedis) from Variable column using regex

```python
money_supply["Variable"] = money_supply["Variable"].str.replace(r"\(Millions.*?\)"," ", regex=True)
```

```python
money_supply.head()
```

### Check data types 

```python
money_supply.dtypes
```

### Remove Years before 2017

```python
money_supply = money_supply[money_supply["Year"]>=2017]
```

```python
money_supply.head()
```

### Reshape the dataset by converting months into rows

```python
money_supply=money_supply.melt(
    id_vars=["Year","Variable"],
    value_vars=["January","Febuary","March","April","May","June","July","August","September","October","November","December"],
    var_name="Month",
    value_name="Value"
)
```

```python
money_supply
```

### Check if there are missing values

```python
money_supply.isnull().sum()
```

### Check if any empty value exists

```python
money_supply.isnull().values.any()
```

### Count the total number of zero

```python
(money_supply["Value"]==0).sum()
```

### View rows with zeros values

```python
money_supply[money_supply["Value"]==0]
```

### Group by variable

```python
money_supply[money_supply["Value"]==0].groupby("Variable").size()
```

### Check if zeros cluster in specific months

```python
money_supply[money_supply["Value"]==0].groupby("Month").size()
```

### Remove Duplicate

```python
money_supply = money_supply.drop_duplicates()
```

### Final dataset Preview

```python
money_supply.head()
```

### Export clean data

```python
money_supply.to_excel(r"C:\Users\TOSHIBA\Desktop\py\Money Supply.xlsx", index=False)
```

### Clean the columns

```python
mobile_money.rename(columns={
    "countrynewwb":"Country",
    "codewb":"Code",
    "year":"Year",
    "pop_adult":"Population_Adult",
    "regionwb24_hi":"Region_Group",
    "incomegroupwb24":"Income_Group",
    "group":"Group_Type",
    "group2":"Demographic_Group",
    "account_t_d": "Account_Owner",
    "fiaccount_t_d": "Financial_Account",
    "mobileaccount_t_d": "Mobile_Account",
    "borrow_any_t_d":"Borrowed_Account"
}, inplace =True)
```

### Preview

```python
mobile_money
```

### Drop unnecessary columns

```python
mobile_money.drop(columns=["Code","Income_Group","Region_Group"], inplace= True)
```

```python
mobile_money.head()
```

### Filter it to the  year range needed

```python
mobile_money=mobile_money[mobile_money["Year"]>=2017].reset_index(drop=True)
```

```python
mobile_money.head()
```

### Remove fully empty columns &  duplicates

```python

mobile_money = mobile_money.dropna(axis=1, how='all')


mobile_money = mobile_money.drop_duplicates()
```

### Fix inconsistent text values

```python
mobile_money[["Group_Type","Demographic_Group"]] = mobile_money[["Group_Type","Demographic_Group"]].replace({
    "all":"All",
    "income":"Income",
    "gender":"Gender",
    "men":"Men",
    "women":"Women",
    "richest 60%":"Richest 60%",
    "poorest 40%":"Poorest 40%",
    "age_cat":"Age_Category",
    "laborforce":"Labor Force",
    "in laborforce":"In Labor Force",
    "out of laborforce":"Out of Labor Force",
    "secondary edu or more":"Secondary  Education / Higher ",
    "prim edu or less":"Primary Education / Lower",
    "age 25+":"Adults (25+)",
    "ages 15-24":"Youth (15-24)",
    "urbanicity":"Urbanicity",
    "education":"Education",
    "urban":"Urban",
    "rural":"Rural"
},)
```
